```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

```python
#Reading the data
df1 = pd.read_csv('resale-flat-prices-based-on-approval-date-1990-1999.csv')
df2 = pd.read_csv('resale-flat-prices-based-on-approval-date-2000-feb-2012.csv')
df3 = pd.read_csv('resale-flat-prices-based-on-registration-date-from-march-2012-onwards.csv')
```

```python
#Data cleansing
frames = [df1, df2, df3]
df = pd.concat(frames)

df['month'] = pd.to_datetime(df['month'])
df['year'] = df['month'].dt.year

df[['month', 'year', 'town', 'flat_type', 'block', 'street_name', 'storey_range', 'floor_area_sqm', 'flat_model', 'lease_commence_date', 'resale_price']]
```

```python
#Price v Year
ax = (df.pivot_table(values = 'resale_price', index = 'year', aggfunc = np.mean)
     .plot(figsize = (12,5), kind = 'line'))
```
